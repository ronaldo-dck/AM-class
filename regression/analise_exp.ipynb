```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Supondo que as variáveis que você deseja plotar sejam 'variavel_x' e 'variavel_y'
sns.lineplot(x='Square_Footage', y='House_Price', data=df)

# Adicionar título e rótulos
plt.xlabel('Square_Footage')
plt.ylabel('House_Price')

# Mostrar o gráfico
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt

# Dividindo os dados em treino e teste
X_main = df['Square_Footage'].values.reshape(-1, 1)  # Principal variável
y = df['House_Price'].values  # Target

X_train, X_test, y_train, y_test = train_test_split(X_main, y, test_size=0.2, random_state=42)

# Lista de modelos a serem testados
models = {
    'Linear Regression': LinearRegression(),
    'KNN': KNeighborsRegressor(n_neighbors=5),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'SVR': SVR(kernel='linear', C=100),
    'MLP': MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42)
}

# Dicionário para armazenar as previsões
predictions = {}

```

```python

# Criando uma figura com uma matriz 3x2 de subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
axes = axes.flatten()  # Facilita a iteração sobre os eixos


# Ajustando e prevendo para cada modelo
for ax, (model_name, model) in zip(axes, models.items()):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
        
    # Diferenciando entre modelos que geram linhas e aqueles que geram pontos
    if model_name in ('KNN', 'GB', 'RF'):
        ax.scatter(X_test, y_test, label='Original')  # Dados originais
        ax.scatter(X_test, y_pred, color='orange', label=f'Previsão - {model_name}')  # Previsão como pontos
    else:
        ax.scatter(X_test, y_test, label='Original')  # Dados originais
        ax.plot(X_test, y_pred, color='orange', label=f'Previsão - {model_name}', linestyle='-')  # Previsão como linha
        
    ax.set_title(model_name)
    ax.set_xlabel('Square_Footage')
    ax.set_ylabel('House_Price')
    ax.legend()

# Ajustando o layout
plt.tight_layout()
plt.show()
```
